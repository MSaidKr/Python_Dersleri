```python
import random

# Generate a random integer between 1 and 10 (inclusive)
rand_int = random.randint(1, 10)
print(f"Random integer between 1 and 10: {rand_int}")

# Generate a random float between 0.0 and 1.0 (exclusive of 1.0)
rand_float = random.random()
print(f"Random float between 0.0 and 1.0: {rand_float}")

# Generate a random float within a specified range
rand_uniform = random.uniform(5.0, 10.0)
print(f"Random float between 5.0 and 10.0: {rand_uniform}")
```

```python
import random

sayi = random.randint(1,10)
puan = 100
hak = int(input("kaç hak istersiniz? "))
sayac = 0

print('Sayınız 1 ile 10 arasında. Tahmin yapmaya başlayabilirsiniz.')

while hak > 0:
  puan = 100 - sayac * 20
  sayac +=1
  hak -=1
  guess = int(input())
  if (hak == 0):
    print("başaramadın bir daha dene")
  elif (sayi > guess):
    print('sayı daha büyük')
  elif (sayi < guess):
    print('sayı daha küçük')
  else:
    print(f'tebrikler sayıyı {sayac}. denemede buldunuz. Puanınız: {puan}')
    break
```

```python
import random

# Shuffle a list in place
numbers = [1, 2, 3, 4, 5]
print(f"Original list: {numbers}")
random.shuffle(numbers)
print(f"Shuffled list: {numbers}")
```

Aşağıdaki kod örnekleri `random.randint` ve `random.choice` arasındaki farkı göstermektedir:

```python
import random

# random.randint(a,b) Belirli bir aralıkta (minimum ve maksimum değerler arasında) rastgele bir tam sayı üretmek istediğinizde kullanırsınız.

# random.randint örneği: 1 ile 10 arasında rastgele bir tam sayı
random_integer = random.randint(1, 10)
print(f"random.randint(1, 10) sonucu: {random_integer}")

# random.choice örneği: Bir listeden rastgele bir öğe
my_list = ['Kırmızı', 'Mavi', 'Yeşil', 'Sarı']
random_element = random.choice(my_list)
print(f"random.choice(['Kırmızı', 'Mavi', 'Yeşil', 'Sarı']) sonucu: {random_element}")
```

```python
import random
#Zaten mevcut olan bir koleksiyonun (liste, demet vb.) içindeki elemanlardan birini rastgele olarak seçmek istediğinizde kullanırsınız.
# random.choice ile sayı listesinden seçim yapma
number_list = [10, 20, 30, 40, 50]
random_number = random.choice(number_list)
print(f"Sayı listesinden seçilen rastgele eleman: {random_number}")

# random.choice ile farklı türde elemanları olan bir listeden seçim yapma
mixed_list = ['elma', 5, True, 3.14, 'muz']
random_mixed_element = random.choice(mixed_list)
print(f"Karışık listeden seçilen rastgele eleman: {random_mixed_element}")
```

```python
import random

# Choose a random element from a list
my_list = ['apple', 'banana', 'cherry', 'date']
random_element = random.choice(my_list)
print(f"Random element from the list: {random_element}")

# Choose multiple unique random elements from a list (sampling without replacement)
random_sample = random.sample(my_list, 2)
print(f"Random sample of 2 elements: {random_sample}")
```
