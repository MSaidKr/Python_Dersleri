# **Math Modülü Uygulamaları**

```python
import math

value = dir(math)
value = help(math)
value = math.sqrt(49)
value = math.factorial(4)
value = math.floor(5.9)
value = math.ceil(5.9) #ceil = tavan
print(value)
```

```python
import math as m

value = m.sqrt(49)
print(value)
```

```python
from math import sqrt,ceil as m

value = m(64)
print(value)
```
