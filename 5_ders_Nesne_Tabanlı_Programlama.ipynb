```python
class Person:
  tür = 'insan'
  def __init__(self, age, name = 'insanoğlu'):
    self.name = name
    self.age = age
    print('insan oluşturuldu')
  def selamlama(self):
    print(f'merhaba benim adım {self.name}, {self.age} yaşındayım ve benim türüm: {self.tür}')
  def dogum(self):
    print(f'ben {self.name} ve {2026 - (self.age)} yılında doğdum')


insan1 = Person(20, 'Lale')
insan2 = Person(23, "Sinem")
insan3 = Person(22)

print(f'{insan1.name} isimli birey, {insan1.age} yaşında. türü: {insan1.tür}')
print(f'{insan2.name} isimli birey, {insan2.age} yaşında. türü: {insan2.tür}')
insan1.selamlama()
insan2.selamlama()
insan1.dogum()
insan2.dogum()
insan3.selamlama()
```

# **Alt Sınıf Oluşturma, Override Etme ve Super()**

```python
class Human():
  def __init__(self,name,lname):
    self.name = name
    self.lname = lname
    print('human created')

  def kimbu(self):
    print(f'benim adım {self.name} {self.lname}')

  def eat(self):
    print(f'çok şükür bugün de doyduk -{self.name}')


class Student(Human):
  def __init__(self,name,lname,number=0):
    self.number = number
    Human.__init__(self,name,lname)
    print('student created')

  def kimbu(self):
    print(f'ben {self.name} {self.lname}, {self.number} numaralı öğrenciyim')


class Teacher(Human):
  def __init__(self,name,lname,branch):
    self.branch = branch
    super().__init__(name,lname)
    print('teacher created')

  def kimbu(self):
    print(f'dikkat {self.branch} hocanız geldi -{self.name} {self.lname}')





Human1 = Human('said','karaahmetoglu')
print(Human1.name + ' ' + Human1.lname)
Human1.kimbu()
Human1.eat()
print('********************************')

human2 = Student('ali','kaya',55)
print(human2.name)
human2.eat()
human2.kimbu()
print('********************************')

human3 = Student('veli','gelgellioğlu')
human3.kimbu()
print('********************************')

human4 = Teacher('ahmet','koca','Robotik')
human4.kimbu()
```

```python
class animal():
  def __init__(self,tur,cinsiyet):
    self.tur = tur
    self.cinsiyet = cinsiyet

  def nedir(self):
    print(f'bu bir {self.tur}')

class dog(animal):
  def __init__(self,tur,cinsiyet,ses):
    animal.__init__(self,tur,cinsiyet)
    self.ses = ses
  def seslendir(self):
    print(f'{self.ses}')

class cat(animal):
  def __init__(self,tur,cinsiyet,ses='mrrrrr'):
    self.ses = ses
    super().__init__(tur,cinsiyet)
  def seslendir(self):
    print(f'{self.ses}')


class bird(animal):
  def __init__(self,tur,cinsiyet,ses):
    self.ses = ses
    super().__init__(tur,cinsiyet)
  def seslendir(self):
    print(f'{self.ses}')




hayvan1 = animal('kedi','erkek')
hayvan1.nedir()
print('********************************')

hayvan2= dog('husky','erkek','hav hav')
hayvan2.nedir()
hayvan2.seslendir()
print('********************************')

hayvan3 = cat('persian','erkek')
hayvan3.nedir()
hayvan3.seslendir()
print('********************************')


hayvan4 = bird('civciv','kız','cik cik')
hayvan4.nedir()
hayvan4.seslendir()
```

# **Quiz Uygulaması - Gem**

```python
class question():
  def __init__(self,text,choices,answer):
    self.text = text
    self.choices = choices
    self.answer = answer

  def checkanswer(self, user_answer):
    # Cevapları büyük/küçük harf duyarlılığı olmadan kontrol et
    return self.answer.lower() == user_answer.lower()

# Soruları tanımla
q1 = question('En kolay programlama dili hangisidir?', ['C++', 'Python', 'Java', 'GDScript'], 'Python')
q2 = question('En iyi meyve hangisidir?', ['Elma', 'Armut', 'Ananas', 'Üzüm'], 'Ananas')
q3 = question('2x + 6 = 8, x= ?', ['1', '2', '3', '4'], '1')

# Soruları bir liste içinde topla
qlist = [q1, q2, q3]

# Quiz başlatma ve yönetme mekanizması (sınıf olmadan)
def start_quiz_without_class(questions):
  score = 0
  total_questions = len(questions)

  for i, current_q in enumerate(questions):
    print(f'{total_questions} sorudan, {i + 1}. soru'.center(100,'*'))
    print(f'{i + 1}. soru: {current_q.text}')
    for j, choice in enumerate(current_q.choices):
      print(f'{j + 1}. seçenek: {choice}')

    user_input = input('Seçiminiz (cevap metni veya seçenek numarası): ').strip()
    is_correct = False

    if user_input.isdigit(): # Kullanıcı sayı girmişse, seçenek numarasını kontrol et
      try:
        choice_index = int(user_input) - 1
        if 0 <= choice_index < len(current_q.choices):
          if current_q.checkanswer(current_q.choices[choice_index]):
            is_correct = True
        else:
          print('Geçersiz seçenek numarası girdiniz.')
      except ValueError:
        print('Geçersiz giriş.')
    else: # Kullanıcı metin girmişse, direkt cevabı kontrol et
      if current_q.checkanswer(user_input):
        is_correct = True

    if is_correct:
      score += 1
      print('Cevabınız DOĞRU!')
    else:
      print(f'Cevabınız YANLIŞ. Doğru cevap: {current_q.answer}')

    print(f'Mevcut Puanınız: {score}')
    print('=' * 100) # Bir sonraki soru için ayırıcı

  print('QUIZ BİTTİ!'.center(100, '='))
  print(f'Toplam Puanınız: {score} / {total_questions}')

# Quiz uygulamasını başlat
start_quiz_without_class(qlist)
```

# **QUIZ UYGULAMASI**
