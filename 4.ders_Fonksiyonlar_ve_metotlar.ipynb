## **Dictionary alıştırmaları****

```python
def consoleUser(**a):
  for key,value in a.items():
    print(f'{key} : {value}')

consoleUser(name = 'user',city = 'istanbul',age = 22)
```

```python
def parktakiler(**a):
  for baslık,icerik in a.items():
    print(f'{baslık} : {icerik}')

parktakiler(Isim = 'Lale',sehir = 'istanbul')
```

```python
def sozluk(**icerik):
  for ad,acıklama in icerik.items():
    print(f'Kelime: {ad} - Açıklama: {acıklama}')
    print(type(icerik))

sozluk(elma = 'amasyadan geldi sulu sulu',armut = 'deveci armutu, yerlidir')
```

```python
def said(**yargi):
  for sey,dusunce in yargi.items():
    print(f'yargılanan: {sey} / yargı: {dusunce}')
said(ampul = 'yanmıyor',ısırgan = 'şerro',said = 'çok müq bir şey')
```

```python
print('ilk parametre :')
x = float(input())
print('ikinci parametre :')
y = float(input())

def uzaklık(**a):
  for parametre1,parametre2 in a.items():
    # Hata düzeltme: float'ları string'e dönüştürmek için f-string kullanıldı
    print(f'{parametre1} {parametre2}')

uzaklık(parametre1 = x , parametre2 = y)
uzaklık = (x*x + y*y)**(1/2)
print(f'merkeze uzaklık: {uzaklık}')
```

```python
def listele(*params): #istediğimiz elemanları listeleyen fonksiyon
  liste = []

  for param in params:
    liste.append(param)
  return liste

result = listele(1,2,3,4,5,6,7,8,9,'lo')
print(result)
```

```python
def ekleyici(*eklenenler):
  liste = []
  for eleman in eklenenler:
    liste.append(eleman)
  return liste

a = ekleyici(12,32,2,3,1,'lo','laa')
print(a)
```

```python
sayi1 = int(input('ilk sayıyı giriniz: '))
sayi2 = int(input('ikinci sayıyı giriniz: '))

def asalcı(sayi1,sayi2):
  for sayi in range(sayi1,sayi2+1):
    if sayi > 1:
      for i in range(2,sayi):
        if (sayi % i) == 0:
          break
      else:
        print(sayi)

asalcı(sayi1,sayi2)
```

```python
def bolenci(sayi):
  bolenler = []
  for i in range(1,sayi+1):
    if sayi % i == 0:
      bolenler.append(i)
  print(bolenler)
bolenci(20)
```

```python
def square(sayi):
  return sayi ** 2

numbers = [1,2,3,4,5]

result = list(map(square,numbers))
print(result)

for i in numbers:
  print(square(i))

for item in map(square,numbers):
  print(item)
```

# **LAMBDA EXPRESSIONS - LAMBDA İFADELERİ**

```python
numbers = [2,3,4]
result = list(map(lambda sayi: sayi**2,numbers)) #anlık fonksiyon tanımlama
print(result)
```

```python
tekleyici = lambda sayi1 : sayi1 + (sayi1 + 1)
numeros = [1,3,5,7,8]
result = list(map(tekleyici,numeros))
print(result)
```

```python
ciftmi = lambda sayi : sayi % 2 == 0
print(ciftmi(2))
```

```python
asalmi = lambda sayi : all(sayi % i != 0 for i in range(2,sayi))
asalmi(17)
```

```python
asalmi2 = lambda sayi : not(any(sayi % i == 0 for i in range(2,sayi)))
asalmi2(17)
```

**Global Değişken Atama**

```python
x = 50
def xfonk():
  x=100
  print(f'x fonksiyonu içindeki x değeri: {x}')

xfonk()
print(x)
#değişmeyen hali
```

```python
x=50
def xchanger():
   global x
   x = 100
   print(f'x fonksiyonu içindeki x değeri: {x}')
   print('x başarıyla değiştirildi.')

xchanger()
print(x)
#global değişkeni değiştiren fonksiyon
```

# **BANKAMATİK UYGULAMASI**

```python
hesap2 = {
    'ad':'muhammed',
    'hesapno':'000888',
    'bakiye':6000,
    'ekhesap':1000
}



hesap1 = {
    'ad':'said',
    'hesapno':'555333',
    'bakiye':3000,
    'ekhesap':5000
}


def paracek(hesap,miktar):
  print(f"merhaba {hesap['ad']}")

  if hesap['bakiye'] >= miktar:
    print('paranızı alabilirsiniz')
    hesap['bakiye'] -= miktar
    print(f"{hesap['hesapno']} nolu hesabınızda {hesap['bakiye']} TL kaldı.")
  else:
    toplam = hesap['bakiye'] + hesap['ekhesap']
    if toplam >= miktar:
      ekhesapmiktar = input('ek hesapdan para çekmek istiyor musunuz? (e/h)')
      if ekhesapmiktar == 'e':
        ekhesapdanmiktar = miktar - hesap['bakiye']
        hesap['bakiye'] = 0
        hesap['ekhesap'] -= ekhesapdanmiktar
        print('paranızı alabilirsiniz')
        print(f" {hesap['hesapno']} nolu hesabınızda {hesap['bakiye']} TL bakiye \nve ekhesabınızda {hesap['ekhesap']} TL bakiye kaldı kaldı.")
      if ekhesapmiktar == 'h':
        print(f"{hesap['hesapno']} nolu hesabınızda {hesap['bakiye']} TL var. Eksik miktar: {miktar - hesap['bakiye']}")
    else:
      print('yetersiz bakiye')

paracek(hesap2,6500)
paracek(hesap2,6500)
```
